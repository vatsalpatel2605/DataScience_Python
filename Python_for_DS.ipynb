```python
#Loop entry for Stacked histogram
list1 = []
mylabels = []
for gen in df1.Genre.cat.categories:
    list1.append(df1[df1.Genre == gen].BudgetMillions)
    mylabels.append(gen)

plt.hist(list1, stacked=True, bins=30, rwidth = 1, label=mylabels)
plt.legend()
plt.show()
```
